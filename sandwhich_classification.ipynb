```python
from fastai.vision import *
from fastai.metrics import error_rate
```

```python
bs=16
```

```python
data_images = ImageDataBunch.from_folder("dataset", test="test", size=(384, 384), bs=bs)
```

```python
data_images.show_batch(rows=3)
```

```python
learn = cnn_learner(data_images, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4, max_lr=slice(1e-6, 1e-4))
```

```python

data_images.test_ds.x[0]
test_img = open_image("C:/Users/ang79722/Documents/fastai_tests/dataset/test/hamburger/3550428.jpg")
```

```python
test_img.show()
```

```python
learn.predict(test_img)
```

```python
preds, _ = learn.get_preds(ds_type=DatasetType.Test)
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.plot_confusion_matrix()
```
